```python
import pandas as pd
import json
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#df = pd.read_json('keyword_counts.json')
with open('keyword_counts.json','r', encoding='utf-8') as f:
    data = json.loads(f.read())
    
# Flatten data
df = pd.json_normalize(data, record_path =['result'])
```

```python
column_names = df.columns
df=df.reindex(columns=['title', 'referencedByKeyword'])
df = df.rename(columns={'title':'Keyword','referencedByKeyword':'Count'})
df.sort_values(by=['Count'], ascending=False, inplace=True)
```

```python
print(df['Count'].sum())
```

```python
display(df.loc[(df['Count'] <= 40)])
```
